```python
# IP-мекенжайлармен жұмыс істеуге арналған кітапхананы импорттау
import ipaddress

# Берілген IP-мекенжайлар
ip1 = ipaddress.IPv4Address('130.20.27.8')  # Бірінші IP-мекенжай
ip2 = ipaddress.IPv4Address('130.28.30.28')  # Екінші IP-мекенжай

# /14 желі маскасы (255.252.0.0) 130.20.0.0 - 130.23.255.255 диапазонын қамтиды
netmask = ipaddress.IPv4Network('130.20.0.0/14', strict=False)

# Екі IP бір желіге тиесілі ме, тексереміз
same_subnet = ip1 in netmask and ip2 in netmask  # Екі IP мекенжайдың да желі ішінде болуын тексереміз

# Нәтижені шығару
print("Екі IP-мекенжай бір желіде орналасқан." if same_subnet else "IP-мекенжайлар әртүрлі желілерде орналасқан.")
# Егер екі IP бір желіге тиесілі болса, тиісті хабарлама шығарылады, әйтпесе басқа хабарлама басылады

```

```python
import ipaddress

# Берілген IP-мекенжайлар
ip1 = int(ipaddress.IPv4Address('130.20.27.8'))   # Бірінші IP бүтін санға айналады
ip2 = int(ipaddress.IPv4Address('130.28.30.28'))  # Екінші IP бүтін санға айналады

# /14 желі маскасы (255.252.0.0)
netmask = int(ipaddress.IPv4Address('255.252.0.0'))  # Масканы да бүтін санға айналдырамыз

# Биттік AND операциясын орындау
subnet1 = ip1 & netmask  # Бірінші IP үшін желі идентификаторы
subnet2 = ip2 & netmask  # Екінші IP үшін желі идентификаторы

# Екі IP бір желіде ме, тексереміз
same_subnet = subnet1 == subnet2  # Егер желі идентификаторлары тең болса, онда IP-мекенжайлар бір желіде

# Нәтижені шығару
print("Екі IP-мекенжай бір желіде орналасқан." if same_subnet else "IP-мекенжайлар әртүрлі желілерде орналасқан.")
```

```python
#24 
import ipaddress

# ✅ Желіні анықтаймыз: 200.100.0.0/22
# Бұл дегеніміз, бізде 22 биттік маска бар, яғни желі бөлігі 22 битті құрайды, қалған 10 бит – хосттар үшін.
network = ipaddress.IPv4Network("200.100.0.0/22", strict=False)

# ✅ Хосттарға қолдануға болатын IP мекенжайлардың санын есептейміз
# Барлық IP мекенжайлардың жалпы саны: 2^(32 - 22) = 2^10 = 1024
# Желілік (network) және кең тарату (broadcast) мекенжайларын қоспағанда, қолдануға болатын мекенжайлар:
num_hosts = network.num_addresses - 2  

# ✅ Бірінші және соңғы хост мекенжайларын анықтаймыз
# Бірінші хост – желінің бірінші мекенжайынан кейінгі мекенжай (network_address + 1)
first_host = network.network_address + 1  

# Соңғы хост – кең тарату мекенжайынан (broadcast_address) бір кем (broadcast_address - 1)
last_host = network.broadcast_address - 1  

# ✅ Нәтижелерді шығару
print(f"Қолдануға болатын хост мекенжайларының саны: {num_hosts}")
print(f"Хосттар диапазоны: {first_host} — {last_host}")

```

```python
#33 
import ipaddress

# Бастапқы желі
network = ipaddress.IPv4Network("10.1.0.0/16", strict=False)

# Кіші желілерге бөлу үшін жаңа маска
new_prefix = 21  # /21 маскасы 32 кіші желіге бөледі

# Жаңа кіші желілерді алу
subnets = list(network.subnets(new_prefix=new_prefix))

# Алғашқы 18 кіші желінің ақпаратын шығару
print("№  | Желі адресі  | Бірінші хост    | Соңғы хост     | Broadcast")
print("---|--------------|----------------|---------------|-------------")

for i, subnet in enumerate(subnets[:18], start=1):
    first_host = subnet.network_address + 1  # Бірінші қолданылатын IP
    last_host = subnet.broadcast_address - 1  # Соңғы қолданылатын IP
    print(f"{i:2} | {subnet} | {first_host} | {last_host} | {subnet.broadcast_address}")
```

```python
#43 
import ipaddress

# Бастапқы желі
network = ipaddress.IPv4Network("200.200.64.0/18", strict=False)

# Әр кіші желіде кемінде 500 IP мекенжай болу керек
# Хосттардың максималды саны формула бойынша: 2^(32 - маска) - 2
# 2^N - 2 ≥ 500 -> 2^N ≥ 502 -> N = 9 (512 IP мекенжай)

new_prefix = 23  # /23 маскасы 512 мекенжай береді (510 қолдануға болады)

# Жаңа кіші желілерді алу
subnets = list(network.subnets(new_prefix=new_prefix))

# Барлық кіші желілерді шығару
print("№  | Желі адресі   | Бірінші хост    | Соңғы хост     | Broadcast     | Маска")
print("---|---------------|----------------|---------------|--------------|-------------")

for i, subnet in enumerate(subnets, start=1):
    first_host = subnet.network_address + 1  # Бірінші қолданылатын IP
    last_host = subnet.broadcast_address - 1  # Соңғы қолданылатын IP
    print(f"{i:2} | {subnet.network_address} | {first_host} | {last_host} | {subnet.broadcast_address} | {subnet.netmask}")
```

```python
#33
import ipaddress  

n = ipaddress.IPv4Network("10.1.0.0/16", strict=False)  

for i, s in enumerate(n.subnets(new_prefix=21), 1):
    print(i, s, s[1], s[-2], s[-1])  # Нөмір, желі, 1-хост, соңғы хост, broadcast  
    if i == 18:
        break
```
